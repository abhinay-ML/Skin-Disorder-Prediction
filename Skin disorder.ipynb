## MODEL CREATION

```python

clinical_features = df.columns[0:11]        
histopath_features = df.columns[11:33]      
all_features = df.columns[0:34]             
target_col = df.columns[-1]                 

X_clinical = df[clinical_features]
X_histopath = df[histopath_features]
X_combined = df[all_features]
y = df[target_col]
```

```python
from sklearn.model_selection import train_test_split

def stratified_split(X, y):
    return train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

Xc_train, Xc_test, yc_train, yc_test = stratified_split(X_clinical, y)
Xh_train, Xh_test, yh_train, yh_test = stratified_split(X_histopath, y)
Xa_train, Xa_test, ya_train, ya_test = stratified_split(X_combined, y)
```

```python
from sklearn.preprocessing import StandardScaler

scaler_c = StandardScaler()
scaler_h = StandardScaler()
scaler_a = StandardScaler()

Xc_train_s = scaler_c.fit_transform(Xc_train)
Xc_test_s = scaler_c.transform(Xc_test)

Xh_train_s = scaler_h.fit_transform(Xh_train)
Xh_test_s = scaler_h.transform(Xh_test)

Xa_train_s = scaler_a.fit_transform(Xa_train)
Xa_test_s = scaler_a.transform(Xa_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neural_network import MLPClassifier

rf = RandomForestClassifier(random_state=42)
svm = SVC(kernel='rbf', random_state=42)
mlp = MLPClassifier(hidden_layer_sizes=(64, 32), max_iter=500, random_state=42)
```

```python
from sklearn.metrics import accuracy_score, f1_score

def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred, average='macro')
    return acc, f1
```

```python
results = []

# Clinical
results.append(["Clinical", "Random Forest", *evaluate_model(rf, Xc_train, Xc_test, yc_train, yc_test)])
results.append(["Clinical", "SVM", *evaluate_model(svm, Xc_train_s, Xc_test_s, yc_train, yc_test)])
results.append(["Clinical", "Neural Network", *evaluate_model(mlp, Xc_train_s, Xc_test_s, yc_train, yc_test)])

# Histopathological
results.append(["Histopathological", "Random Forest", *evaluate_model(rf, Xh_train, Xh_test, yh_train, yh_test)])
results.append(["Histopathological", "SVM", *evaluate_model(svm, Xh_train_s, Xh_test_s, yh_train, yh_test)])
results.append(["Histopathological", "Neural Network", *evaluate_model(mlp, Xh_train_s, Xh_test_s, yh_train, yh_test)])

# Combined
results.append(["Combined", "Random Forest", *evaluate_model(rf, Xa_train, Xa_test, ya_train, ya_test)])
results.append(["Combined", "SVM", *evaluate_model(svm, Xa_train_s, Xa_test_s, ya_train, ya_test)])
results.append(["Combined", "Neural Network", *evaluate_model(mlp, Xa_train_s, Xa_test_s, ya_train, ya_test)])
```

```python
import pandas as pd

# Convert list of results into a DataFrame
results_df = pd.DataFrame(results, columns=["Feature_Set", "Model", "Accuracy", "F1_Score"])

# Round scores for readability
results_df["Accuracy"] = results_df["Accuracy"].round(3)
results_df["F1_Score"] = results_df["F1_Score"].round(3)

# Display neatly
print("Model Performance Summary:\n")
print(results_df.to_string(index=False))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def evaluate_model(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, average='macro')
    rec = recall_score(y_test, y_pred, average='macro')
    f1 = f1_score(y_test, y_pred, average='macro')
    
    return acc, prec, rec, f1
```

```python
results = []

# Clinical
results.append(["Clinical", "Random Forest", *evaluate_model(rf, Xc_train, Xc_test, yc_train, yc_test)])
results.append(["Clinical", "SVM", *evaluate_model(svm, Xc_train_s, Xc_test_s, yc_train, yc_test)])
results.append(["Clinical", "Neural Network", *evaluate_model(mlp, Xc_train_s, Xc_test_s, yc_train, yc_test)])

# Histopathological
results.append(["Histopathological", "Random Forest", *evaluate_model(rf, Xh_train, Xh_test, yh_train, yh_test)])
results.append(["Histopathological", "SVM", *evaluate_model(svm, Xh_train_s, Xh_test_s, yh_train, yh_test)])
results.append(["Histopathological", "Neural Network", *evaluate_model(mlp, Xh_train_s, Xh_test_s, yh_train, yh_test)])

# Combined
results.append(["Combined", "Random Forest", *evaluate_model(rf, Xa_train, Xa_test, ya_train, ya_test)])
results.append(["Combined", "SVM", *evaluate_model(svm, Xa_train_s, Xa_test_s, ya_train, ya_test)])
results.append(["Combined", "Neural Network", *evaluate_model(mlp, Xa_train_s, Xa_test_s, ya_train, ya_test)])
```
